````python
import pandas as pd
````

#### Loading Data

````python
DATA_DIR = 'data'
XL_FILE = 'DSS BrainPost Web Analytics Challenge!.xlsx'
````

###### Let's try to understand the structure in the data file 
Objective is to know:
    - how many sheets we have,
    - what data is there in each sheet,
    - which information we need for analysis and which data is repeated.
    
So let's start looking into the file:

First of all, let's check the sheets,

````python
xls = pd.ExcelFile(f"{DATA_DIR}/{XL_FILE}")
for sheet_name in xls.sheet_names:
    print(sheet_name)
````

The first three sheets are for information, That we already have read. Let's chekc the 4th sheet.

````python
xl_df = xls.parse(xls.sheet_names[8])
xl_df.head(10)
````

Looks like we have some comments on top. Let's skip them.

````python
xl_df = xls.parse(xls.sheet_names[8],  skiprows=6)
xl_df.head()
````

Now it looks good. What about the tail? checking 25 last entries wiht ```xl_df.tail(25)``` showd there is a different data at the bottom.

````python
xl_df.tail(25)
````

We need to skip last 20 rows as we don't need sum of the first table as it is in second table too.

````python
xl_df = xls.parse(xls.sheet_names[8],  skiprows=6, skipfooter=20)
xl_df.tail()
````

Now it looks perfect. But what is the data in last 20 rows. It looks like the page view counts of different durations. Which can be taken from above dataframe. But just in case, we need the sum with those ranges of date. Let's keep that data too. but in a separate dataframe.
We will need some cleanup here too. 
- That means to clean up we need to skip the  comments on top(6 rows) ,rows whole upper table(xl_df.shape[0]+1) +  rows of sums in upper table(2) + blank line in between(1).
- also, we have only three columns in lower table.

Let's do it..

````python
print(xls.sheet_names[8])
page_view_df = xls.parse(xls.sheet_names[8], skiprows=xl_df.shape[0]+1 + 9)
page_view_df = page_view_df.iloc[:, 0:3]
page_view_df
````

- As we see, the lower table ultimately groups the page views from two data ranges so the table itself is not much important as it has ranges repeated as counted on differnt 'day index' value. Let's just keep the sum from bottom.

````python
page_view_df.iloc[-2:,1:] 
````

These two dataframes are we have in this sheet. __But we may not need the page view dataframe as it is just the repeated data that we already have in upper table we extracted in ```xl_df```.__

Similar, analysis done quickly on other sheets shows that all the other sheets have similar structure with different number of the data entries. So we are good to combine the sheets into one dataframe for ease of analysis.
Let's try that...

````python
non_data_sheets = ['About the Challenge','Signup Page','Page to Page Title lookup']
xls = pd.ExcelFile(f"{DATA_DIR}/{XL_FILE}")
full_data_df = pd.DataFrame()
full_page_view_df = pd.DataFrame()
data_sheets = [sheet_name for sheet_name in xls.sheet_names if sheet_name not in non_data_sheets]
for sheet_name in data_sheets:
    content_data_df = xls.parse(sheet_name, skiprows=6, skipfooter=20)
    full_data_df = full_data_df.append(content_data_df, ignore_index=True)
    page_view_df = xls.parse(sheet_name, skiprows=content_data_df.shape[0]+1 + 9)
    full_page_view_df = full_page_view_df.append(page_view_df.iloc[-2:,1:3], ignore_index=True)
````

Now we have two dataframes with all the data we need for further analysis.

````python
print(f'Content data shape: {full_data_df.shape}')
print(f'Page View data shape: {full_page_view_df.shape}')
````

#### Saving extracted dataframes for future
Now, we have two dataframes let's save them for future use.

````python
full_data_df.to_csv(f'{DATA_DIR}/content_data.csv', index=False)
full_page_view_df.to_csv(f'{DATA_DIR}/page_view_data.csv', index=False)
````
